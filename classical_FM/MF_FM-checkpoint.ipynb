```python
import torch 
from torch import nn, Tensor

import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

import ot
import numpy as np
```

```python
class Flow(nn.Module):
    def __init__(self, dim: int = 2, h: int = 64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(dim + 1, h), nn.ELU(),
            nn.Linear(h, h), nn.ELU(),
            nn.Linear(h, h), nn.ELU(),
            nn.Linear(h, dim))
    
    def forward(self, t: Tensor, x_t: Tensor) -> Tensor:
        return self.net(torch.cat((t, x_t), -1))
    
    def step(self, x_t: Tensor, t_start: Tensor, t_end: Tensor) -> Tensor:
        t_start = t_start.view(1, 1).expand(x_t.shape[0], 1)
        
        return x_t + (t_end - t_start) * self(t=t_start + (t_end - t_start) / 2, x_t= x_t + self(x_t=x_t, t=t_start) * (t_end - t_start) / 2)
```

```python
def sample_eight_gaussians(n_samples, radius=1.5, std=0.1, seed=None):
    """
    从八个高斯混合分布中采样。
    八个高斯中心位于半径为 `radius` 的圆上，等间隔角度。
    每个分量的协方差为 std^2 * I。
    """
    if seed is not None:
        np.random.seed(seed)
    angles = np.linspace(0, 2 * np.pi, 8, endpoint=False)
    centers = radius * np.stack([np.cos(angles), np.sin(angles)], axis=1)
    # 为每个样本随机选择一个分量（均匀分布）
    comp_ids = np.random.randint(0, 8, size=n_samples)
    samples = []
    for i in range(n_samples):
        mean = centers[comp_ids[i]]
        sample = mean + std * np.random.randn(2)
        samples.append(sample)
    return np.array(samples, dtype=np.float32)
```

```python
def compute_repulsion_loss(X, sigma=0.1):
    """
    计算当前时间t时刻，Batch内粒子的排斥能损失 (Repulsion Loss)
    
    参数:
    X: 形状为 (B, D) 的张量，表示当前时刻一个Batch内的B个粒子，每个粒子维度为D
    sigma: 核函数的带宽。如果为 None，则自动启用 Median Trick 动态计算
    
    返回:
    loss: 标量张量，可直接用于 loss.backward()
    """
    B, D = X.shape
    # dist_squared[i, j] = ||x_i - x_j||^2
    dist_squared = torch.cdist(X, X, p=2) ** 2
    
    two_sigma_squared = 2 * (sigma ** 2)
        
    # 3. 计算高斯核矩阵 K, 形状为 (B, B)
    K = torch.exp(-dist_squared / two_sigma_squared)
    
    # 4. 提取非对角线元素（因为对角线 i=j 时 K[i,i]=1 是常数，无梯度）
    # 使用 mask 过滤掉对角线
    mask = torch.eye(B, device=X.device, dtype=torch.bool)
    repulsion_elements = K[~mask]
    
    # 5. 我们希望最小化这些核值（核值越小，粒子间距越大，即实现了排斥）
    loss = repulsion_elements.mean()
    
    return loss
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42

np.random.seed(SEED)
torch.manual_seed(SEED)
```

```python
def TD_sampler(target_distribution, BatchSize):
    if target_distribution == '8Gaussian':
        x_1 = sample_eight_gaussians(BatchSize, radius=1.5, std=0.1, seed=SEED)  # numpy array
        return x_1
    elif target_distribution == 'moon':
        x_1 = make_moons(BatchSize, noise=0.15, random_state=SEED)[0]
        return x_1
```

```python
delta_t = 0.01
lambda_1 = 1
sigma = 0.5
outer_loop = 1000
inner_loop = 100
target_distribution = '8Gaussian' # or 'moon'
```

```python
# GPU training
flow = Flow().to(device)
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(flow.parameters(), 1e-4)
losses = []
Loss_1 = []
Loss_2 = []

for _ in range(outer_loop):
    x_1 = torch.tensor(TD_sampler(target_distribution, 256), dtype=torch.float32).to(device)
    # x_1_np = sample_eight_gaussians(256, radius=1.5, std=0.1, seed=SEED)  # numpy array
    # x_1 = torch.tensor(x_1_np, dtype=torch.float32).to(device)
    # x_1 = torch.tensor(make_moons(256, noise=0.15, random_state=SEED)[0], dtype=torch.float32).to(device)
    x_0 = torch.randn_like(x_1).to(device)
    # t   = torch.rand(len(x_1), 1).to(device)
    for __ in range(inner_loop):
        t = torch.rand(1, 1).expand(x_1.size(0), 1).to(device)
        x_t = (1 - t) * x_0 + t * x_1
        v_pred = flow(t=t, x_t=x_t)
        x_t_dt = x_t + delta_t * v_pred
        dx_t = x_1 - x_0
        loss_1 = loss_fn(v_pred, dx_t)
        loss_2 = compute_repulsion_loss(x_t_dt, sigma=sigma)
        loss_total = loss_1 + lambda_1 * loss_2
        optimizer.zero_grad()
        loss_total.backward()
        optimizer.step()
        
        Loss_1.append(loss_1.item())
        Loss_2.append(loss_2.item())
        losses.append(loss_total.item())

# 绘制损失曲线
plt.figure(figsize=(10, 5))
plt.plot(losses)
plt.xlabel('Training Step')
plt.ylabel('Total Loss')
plt.title('Training Loss')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(Loss_1)
plt.xlabel('Training Step')
plt.ylabel('Loss_1')
plt.title('FM Loss')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(Loss_2)
plt.xlabel('Training Step')
plt.ylabel('Loss_2')
plt.title('Repulsion Loss')
plt.grid(True)
plt.show()
```

```python
# %% 观察分布转移路径 (Flow 模型)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

def plot_transfer_paths(model, n_samples=30, n_steps=50, ax=None):
    """
    生成并绘制从噪声到目标分布的样本转移路径。
    
    参数:
        model: 训练好的 Flow 模型（CPU 模式）
        n_samples: 要显示的路径数目
        n_steps: 从 t=0 到 t=1 的积分步数
        ax: 可选，用于绘图的坐标轴
    """
    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 5))
    
    # 1. 生成初始噪声样本 (先验分布为标准高斯)
    x = torch.randn(n_samples, 2)
    t_vec = torch.linspace(0, 1.0, n_steps + 1)
    dt = t_vec[1] - t_vec[0]
    
    # 2. 按时间步逐步积分，记录路径
    # paths[k, i, j] 表示第 k 个时间步，第 i 个样本的第 j 维
    paths = np.zeros((n_steps + 1, n_samples, 2))
    paths[0] = x.detach().cpu().numpy()
    
    with torch.no_grad():
        for s in range(n_steps):
            t_start = t_vec[s]
            t_end   = t_vec[s + 1]
            x = model.step(x_t=x, t_start=t_start, t_end=t_end)
            paths[s + 1] = x.detach().cpu().numpy()
    
    # 3. 绘制路径 (每条路径用渐变色表示时间演化)
    # 为每条样本创建一条折线，并根据时间步着色
    colors = plt.cm.viridis(np.linspace(0, 1, n_steps + 1))  # 颜色从深色(起始)到亮色(终点)
    
    for i in range(n_samples):
        points = paths[:, i, :]  # shape (n_steps+1, 2)
        segments = np.stack([points[:-1], points[1:]], axis=1)
        lc = LineCollection(segments, cmap='viridis', norm=plt.Normalize(0, 1))
        lc.set_array(np.linspace(0, 1, n_steps))
        ax.add_collection(lc)
    
    # 标记起始点和终点
    ax.scatter(paths[0, :, 0], paths[0, :, 1], c='blue', s=30, label='Start (t=0)')
    ax.scatter(paths[-1, :, 0], paths[-1, :, 1], c='red', s=30, label='End (t=1)')
    
    # 生成真实数据作为背景参考
    # real_data = sample_eight_gaussians(1000, radius=1.5, std=0.1, seed=SEED)
    # real_data = make_moons(1000, noise=0.15)[0]
    real_data = TD_sampler(target_distribution, 1000)
    ax.scatter(real_data[:, 0], real_data[:, 1], c='gray', s=5, alpha=0.3, label='Real Data')
    
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_title('Transfer Paths of Flow Model')
    ax.legend(loc='upper right')
    ax.grid(True, linestyle='--', alpha=0.5)
    return ax

# 绘制 Flow 模型的转移路径
plt.figure(figsize=(6, 5))
plot_transfer_paths(flow, n_samples=100, n_steps=8)
plt.tight_layout()
plt.show()
```
